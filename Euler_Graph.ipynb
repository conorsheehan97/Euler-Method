So our derivitive function was given by:
\begin{equation}
\frac{dy}{dx} = x + y + x*y 
\end{equation}
With initial conditions $(x_{0}, y_{0})$ = 0,1 with x ranging from 0 to 1.

Analytically, the above can be integrated to using Wolfram Alpha:
\begin{equation}
 y(x) = \frac{1}{2} e^{(\frac{x^{2}}{2} - \frac{1}{2} (x + 2) x + x)} (\sqrt{2 \pi}(-e^{\frac{1}{2}x(x+2) + \frac{1}{2}}erf(\frac{x+1}{\sqrt{2}})) +  erf(\frac{1}{\sqrt{2}})\sqrt{2\pi}e^{\frac{x(x+2)}{2} + \frac{1}{2}} + 4e^{\frac{x(x+2)}{2}} - 2)
\end{equation}

```python
import matplotlib.pyplot as plt

# Open the text file
with open(r"C:\Users\conor\Documents\C++\Eulers Method\Euler\Euler_data.txt", "r") as file:
    # Read the lines
    lines = file.readlines()

# Initialize empty lists for storing data
column1 = []
column2 = []

# Parse the data
for line in lines:
    # Split the line into columns
    parts = line.strip().split()  # assuming columns are separated by whitespace

    # Extract the values of the two columns
    column1.append(float(parts[0]))
    column2.append(float(parts[1]))

# Plot the data
plt.plot(column1, column2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Euler Method solution for x + y + xy')
plt.grid(True)
plt.show()
```
